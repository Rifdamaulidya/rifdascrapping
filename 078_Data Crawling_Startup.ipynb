```python
import pandas as pd
from bs4 import BeautifulSoup

import chromedriver_autoinstaller
from selenium import webdriver
```

```python
chromedriver_autoinstaller.install()

browser = webdriver.Chrome()
```

```python
browser.get('https://www.forbes.com/lists/americas-best-startup-employers/?sh=6963eceb2ad7')
```

```python
soup = BeautifulSoup(browser.page_source, 'lxml')
```

```python
soup
```

```python
soup.find('div', class_='rank first table-cell rank')
```

```python
soup.find_all('div', class_='table-cell')
```

```python
data_rank = []
data_name =[]
data_industry =[]
data_city =[]
data_state =[]
    
ranks = soup.find_all('div', class_='rank first table-cell rank')
for each_rank in ranks:
    data_rank.append(each_rank.text) 
    
names = soup.find_all('div', class_='organizationName second table-cell name')
for each_name in names:
    data_name.append(each_name.text)

industries = soup.find_all('div', class_='industry table-cell industry')
for each_industry in industries:
    data_industry.append(each_industry.text)
    
cities = soup.find_all('div', class_='city table-cell city')
for each_city in cities:
    data_city.append(each_city.text)

states = soup.find_all('div', class_='state table-cell state')
for each_state in states:
    data_state.append(each_state.text)
```

```python
startup = {
    'Rank': data_rank,
    'Startup Name': data_name,
    'Industry': data_industry,
    'City': data_city,
    'State': data_state,
}
    
pd.DataFrame(startup).to_excel('America Best Startup Employers.xlsx')
```
